```python
from skimage.io import imread
import cv2
import matplotlib.pyplot as plt
```

```python
im_rgb = imread('jelly_beans.tiff')
im_hsv = cv2.cvtColor(im_rgb, cv2.COLOR_BGR2HSV)

mask_rgb = cv2.inRange(im_rgb, (180, 180, 0), (255, 255, 100))
res_rgb = cv2.bitwise_and(im_rgb, im_rgb, mask=mask_rgb)

mask_hsv = cv2.inRange(im_hsv, (85, 135, 185), (95, 255, 255))
res_hsv = cv2.bitwise_and(im_rgb, im_rgb, mask=mask_hsv)

plt.figure(figsize=(30, 15))
plt.subplot(131), plt.imshow(im_rgb), plt.axis('off')
plt.subplot(132), plt.imshow(res_rgb), plt.axis('off')
plt.subplot(133), plt.imshow(res_hsv), plt.axis('off')
plt.show()
```
